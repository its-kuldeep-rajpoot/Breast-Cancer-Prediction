## Model building:

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
X = new_df.drop(['diagnosis'], axis=1)
y = new_df['diagnosis']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

for i in X.columns:
    X[i+' Normalized'] = scaler.fit_transform(X[i].values.reshape(-1, 1))
    X.drop([i], inplace= True, axis= 1)
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.15, stratify=y, random_state=1)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
lr_model = LogisticRegression(max_iter=34, random_state=1)

lr_model.fit(X_train, y_train)
```

```python
y_pred = lr_model.predict(X_test)
```

```python
acc = accuracy_score(y_pred, y_test)
print("Accuracy : ", acc)
```

```python
c = classification_report(y_test, y_pred)
print(c)
```
